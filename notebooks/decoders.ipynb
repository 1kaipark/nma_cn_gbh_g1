# Data Loading


```python
# @title Main Data
# raise Exception("Can only run one cell")
import os, requests, numpy as np


def load_data() -> np.ndarray:
    fname = []
    for j in range(3):
        fname.append("steinmetz_part%d.npz" % j)
    url = ["https://osf.io/agvxh/download"]
    url.append("https://osf.io/uv3mw/download")
    url.append("https://osf.io/ehmw2/download")

    for j in range(len(url)):
        if not os.path.isfile(fname[j]):
            try:
                r = requests.get(url[j])
            except requests.ConnectionError:
                print("!!! Failed to download data !!!")
            else:
                if r.status_code != requests.codes.ok:
                    print("!!! Failed to download data !!!")
                else:
                    with open(fname[j], "wb") as fid:
                        fid.write(r.content)

    main_data = np.array([])
    for j in range(len(fname)):
        main_data = np.hstack(
            (main_data, np.load("steinmetz_part%d.npz" % j, allow_pickle=True)["dat"])
        )
        
    return main_data
  
main_data = load_data()
```

```python
def plot_PCA_activity(
    spks: np.ndarray, resp: np.ndarray, bin_1: int, bin_2: int, title: str
) -> None:
    """plots PCA activity of neural data of shape (n_neurons, n_trials, n_timebins). generates 2 subplots for each bin."""
    X_1 = spks[:, :, bin_1].T
    X_2 = spks[:, :, bin_2].T

    n_neurons, n_trials, n_timebins = spks.shape

    scaler = MinMaxScaler()
    pca = PCA(n_components=3)

    X_scaled_1 = scaler.fit_transform(X_1)
    W_1 = pca.fit_transform(X_scaled_1)
    pc1_1 = W_1[:, 0]
    pc2_1 = W_1[:, 1]
    pc3_1 = W_1[:, 2]

    X_scaled_2 = scaler.fit_transform(X_2)
    W_2 = pca.fit_transform(X_scaled_2)
    pc1_2 = W_2[:, 0]
    pc2_2 = W_2[:, 1]
    pc3_2 = W_2[:, 2]

    response_direction = resp.reshape(n_trials)
    labels_dict = {-1.0: "left", 0.0: "no-go", 1.0: "right"}

    # Create 3D scatter plot
    f = plt.figure(figsize=(10, 5))
    ax_1 = f.add_subplot(1, 2, 1, projection="3d")

    # Plot each response direction with different colors
    for label in np.unique(response_direction):
        indices = response_direction == label
        ax_1.scatter(
            pc1_1[indices], pc2_1[indices], pc3_1[indices], label=labels_dict[label]
        )

    ax_2 = f.add_subplot(1, 2, 2, projection="3d")
    # Plot each response direction with different colors
    for label in np.unique(response_direction):
        indices = response_direction == label
        ax_2.scatter(
            pc1_2[indices], pc2_2[indices], pc3_2[indices], label=labels_dict[label]
        )

    for ax in [ax_1, ax_2]:
        ax.set_xlabel("PC 1")
        ax.set_ylabel("PC 2")
        ax.set_zlabel("PC 3")

    ax_1.set_title("t={} ms".format(bin_1*10))
    ax_2.set_title("t={} ms".format(bin_2*10))
    ax_1.legend(loc="upper left")

    f.suptitle(title)
    f.tight_layout()


BIN_1, BIN_2 = 43, 57

sns.set_style("white")

plot_PCA_activity(
    all_data["spks"][all_data["brain_area"] == "VISp"],
    all_data["response"],
    BIN_1,
    BIN_2,
    "VISp PCA"
)

plot_PCA_activity(
    all_data["spks"][all_data["brain_area"] == "MOs"],
    all_data["response"],
    BIN_1,
    BIN_2,
    "MOs PCA"
)
```

# Logistic Regression


```python
# Helper Functions + Data
def process_spikes(spks: np.ndarray) -> np.ndarray:
    """Intakes spike data of shape (n_neurons, n_trials, n_timebins), sums spike counts across timebins, then returns average over trial duration."""
    n_neurons, n_trials, n_timebins = spks.shape
    total_duration_sec = (n_timebins * 10) / 100  # bin = 10 ms
    total_spike_counts = np.sum(spks, axis=2)  # sum across timebins

    firing_rates = total_spike_counts / total_duration_sec

    norm = Normalizer()
    return norm.fit_transform(firing_rates).T


def fit_regression(X: np.ndarray, y: np.ndarray, params: dict) -> LogisticRegression:
    """Splits data, then fits a given SKLearn linear model. Returns the model"""

    logreg = LogisticRegression(
        tol=params["tol"],
        max_iter=params["max_iter"],
        penalty=params["penalty"],
        solver=params["solver"],
        C=params["C"],
    )
    logreg.fit(X, y)

    return logreg


def plot_confusion_matrix(y_acc: np.ndarray, y_hat: np.ndarray):
    conf_matrix = confusion_matrix(y_acc, y_hat)
    disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)
    disp.plot(cmap=plt.cm.Reds)


def plot_model_accuracies(
    model: LogisticRegression, X: np.ndarray, y: np.ndarray, k: int = 8
):
    accuracies = cross_val_score(
        model,
        X,
        y,
        cv=k,
    )  # k-fold cross validation

    f, ax = plt.subplots(figsize=(8, 3))
    ax.boxplot(accuracies, vert=False, widths=0.7)
    ax.scatter(accuracies, np.ones(8))
    ax.set(
        xlabel="VISp Decoder Accuracy",
        yticks=[],
        title=f"VISp Decoder Average test accuracy: {accuracies.mean():.2%}",
    )
    ax.spines["left"].set_visible(False)


logreg_params = {
    "tol": 0.001,
    "max_iter": 1000,
    "penalty": "l2",
    "solver": "liblinear",
    "C": 1.0,
}

nogo_mask = all_data["response"] != 0
go_response = all_data["response"][nogo_mask]
```

## Visual Decoder

```python
VISp_spikes = all_data["spks"][all_data["brain_area"] == "VISp"]

n_neurons_VISp, n_trials, n_timebins = VISp_spikes.shape
firing_rates_scaled_VISp = process_spikes(VISp_spikes)



go_spikes_VISp = firing_rates_scaled_VISp[nogo_mask]
print(go_spikes_VISp.shape, go_response.shape)  # (trials, neurons)

X_train_VISp, X_test_VISp, y_train, y_test = train_test_split(
    go_spikes_VISp, go_response, train_size=0.8, random_state=0
)

logreg_VISp = fit_regression(X_train_VISp, y_train, logreg_params)

print("Rsquared", logreg_VISp.score(X_test_VISp, y_test))

# Accuracy code
choice_pred_VISp = logreg_VISp.predict(X_test_VISp)  # this gives us y_hat
n_trial_correct_VISp = len(
    np.where(y_test == choice_pred_VISp)[0]
)  # when does y_hat equal y_actual
print(
    f"Accuracy : {n_trial_correct_VISp} trials correctly predicted / {len(y_test)} trials = {n_trial_correct_VISp/len(y_test)}"
)
```

```python
sns.set_style("white")

plot_confusion_matrix(y_test, choice_pred_VISp)
plt.title("VISp Confusion Matrix")
plt.show()

plot_model_accuracies(logreg_VISp, go_spikes_VISp, go_response)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Función para graficar las pérdidas de entrenamiento y validación
def plot_train_val_loss(train_losses, val_losses, title):
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.xlabel('Number of epochs')
    plt.ylabel('Loss')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
AREA = 'VISp'
visual_spks = all_data['spks'][np.ma.mask_or(all_data['brain_area'] == AREA, all_data['brain_area'] == 'VISam')]
# motor_spks = all_data['spks'][all_data['brain_area'] == 'MOs']

bin_spikes_VISp = visual_spks[:, :, 50:100].mean(axis=2)
n_neurons, n_trials = bin_spikes_VISp.shape
X_VISp = bin_spikes_VISp.T

y_variable = all_data['response']

X_excl_nogo_VISp = X_VISp[y_variable != 0]
y_excl_nogo = y_variable[y_variable != 0]
```

```python

# Normalize data
scaler = StandardScaler()
X_input_VISp = scaler.fit_transform(X_excl_nogo_VISp)
X_inputVISp = torch.tensor(X_input_VISp, dtype=torch.float32)

# Dimensionality reduction on data
pca = PCA(n_components=10)
X_input_VISp = pca.fit_transform(X_input_VISp)
print(X_input_VISp.shape)
var_explained_VISp = pca.explained_variance_ratio_
plt.plot(var_explained_VISp)
plt.title("Explained Variance Ratio (VISp)")
```

```python
pc1_VISp = X_input_VISp[:, 0]
pc2_VISp = X_input_VISp[:, 1]
pc3_VISp = X_input_VISp[:, 2]


# Create 3D scatter plot
f = plt.figure(figsize=(5, 5))
ax = f.add_subplot(111, projection='3d')


ax.scatter(pc1_VISp, pc2_VISp, pc3_VISp)

ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
ax.set_zlabel('PC 3')
ax.set_title('VISp PCA')
```

```python
X_tensor_VISp = torch.tensor(X_input_VISp, dtype=torch.float32)
y_acc = torch.tensor(y_excl_nogo, dtype=torch.float32)

X_train_VISp, X_val_VISp, y_train, y_val = train_test_split(X_tensor_VISp, y_acc, test_size=0.2, random_state=42)
```

```python
class DeepNetReLU_1hidden(nn.Module):
    """Network with one hidden layer with ReLU activation"""
    def __init__(self, n_inputs, n_hidden, dropout_rate: float = 0.5):
        super().__init__()
        self.in_layer = nn.Linear(n_inputs, n_hidden)  # Input to hidden layer
        self.out_layer = nn.Linear(n_hidden, 1)  # Hidden layer to output
        self.dropout = nn.Dropout(p=dropout_rate)

    def forward(self, r):
        h = self.in_layer(r)  # h is size (batch_size, n_hidden)
        h = torch.relu(h)

        y = self.out_layer(h)  # y is size (batch_size, 1)
        y = self.dropout(y)

        return y

# Without batch
def train(net, loss_fn, X_train, y_train, X_val, y_val, n_epochs, optimizer_type='adam', lr=0.001):
    if optimizer_type == 'adam':
        optimizer = optim.Adam(net.parameters(), lr=lr)
    elif optimizer_type == 'rmsprop':
        optimizer = optim.RMSprop(net.parameters(), lr=lr)
    elif optimizer_type == 'sgd':
        optimizer = optim.SGD(net.parameters(), lr=lr)
    else:
        raise ValueError("Optimizer not supported")

    train_losses = []
    val_losses = []
    accuracies = []

    for epoch in range(n_epochs):
        net.train()
        optimizer.zero_grad()
        outputs = net(X_train)
        loss = loss_fn(outputs, y_train.unsqueeze(1))
        loss.backward()
        optimizer.step()
        train_loss = loss.item()
        train_losses.append(train_loss)

        net.eval()
        with torch.no_grad():
            val_outputs = net(X_val)
            val_outputs_discrete = np.where(val_outputs < 0, -1, 1).flatten()
            accuracy = (val_outputs_discrete == y_val.detach().numpy()).sum() / len(y_val)
            accuracies.append(accuracy)

            val_loss = loss_fn(val_outputs, y_val.unsqueeze(1)).item()
            val_losses.append(val_loss)

        if (epoch+1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{n_epochs}], Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}')

    return train_losses, val_losses, accuracies


n_hidden = 100
n_neurons_VISp, n_trial_VISp = X_tensor_VISp.T.shape

model = DeepNetReLU_1hidden(n_inputs=n_neurons_VISp, n_hidden=n_hidden, dropout_rate=0.3)

loss_fn = nn.MSELoss()

train_losses_VISp, val_losses_VISp, accuracies_VISp = train(net=model, loss_fn=loss_fn, X_train=X_train_VISp, y_train=y_train, X_val=X_val_VISp, y_val=y_val, n_epochs=80, optimizer_type='adam', lr=0.001)


plot_train_val_loss(train_losses_VISp, val_losses_VISp, title='Visual Cortex Decoder')
```

```python
plt.plot(accuracies_VISp)
plt.title("Visual Cortex: Decoder Accuracy")
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
```

```python
class DeepNetReLU_1hidden(nn.Module):
    """Network with one hidden layer with ReLU activation"""
    def __init__(self, n_inputs, n_hidden, dropout_rate: float = 0.5):
        super().__init__()
        self.in_layer = nn.Linear(n_inputs, n_hidden)  # Input to hidden layer
        self.out_layer = nn.Linear(n_hidden, 1)  # Hidden layer to output
        self.dropout = nn.Dropout(p=dropout_rate)

    def forward(self, r):
        h = self.in_layer(r)  # h is size (batch_size, n_hidden)
        h = torch.relu(h)

        y = self.out_layer(h)  # y is size (batch_size, 1)
        y = self.dropout(y)

        return y

# Without batch
def train(net, loss_fn, X_train, y_train, X_val, y_val, n_epochs, optimizer_type='adam', lr=0.001):
    if optimizer_type == 'adam':
        optimizer = optim.Adam(net.parameters(), lr=lr)
    elif optimizer_type == 'rmsprop':
        optimizer = optim.RMSprop(net.parameters(), lr=lr)
    elif optimizer_type == 'sgd':
        optimizer = optim.SGD(net.parameters(), lr=lr)
    else:
        raise ValueError("Optimizer not supported")

    train_losses = []
    val_losses = []
    accuracies = []

    for epoch in range(n_epochs):
        net.train()
        optimizer.zero_grad()
        outputs = net(X_train)
        loss = loss_fn(outputs, y_train.unsqueeze(1))
        loss.backward()
        optimizer.step()
        train_loss = loss.item()
        train_losses.append(train_loss)

        net.eval()
        with torch.no_grad():
            val_outputs = net(X_val)
            val_outputs_discrete = np.where(val_outputs < 0, -1, 1).flatten()
            accuracy = (val_outputs_discrete == y_val.detach().numpy()).sum() / len(y_val)
            accuracies.append(accuracy)

            val_loss = loss_fn(val_outputs, y_val.unsqueeze(1)).item()
            val_losses.append(val_loss)

        if (epoch+1) % 10 == 0:
            print(f'Epoch [{epoch+1}/{n_epochs}], Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}')

    return train_losses, val_losses, accuracies


n_hidden = 100
n_neurons_con, n_trials_con = X_tensor_con.T.shape

model_con = DeepNetReLU_1hidden(n_inputs=n_neurons_con, n_hidden=n_hidden, dropout_rate=0.3)

loss_fn = nn.MSELoss()

train_losses_con, val_losses_con, accuracies_con = train(net=model_con, loss_fn=loss_fn, X_train=X_train_con, y_train=y_train, X_val=X_val_con, y_val=y_val, n_epochs=400, optimizer_type='adam', lr=0.01)


plot_train_val_loss(train_losses_con, val_losses_con, title='MOs + VISp Concatenated Decoder')
print(accuracies_con)
```
